```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
class MLPVectorField(nn.Module):
    def __init__(self, input_dim=784, hidden_dim=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim + 1, hidden_dim), # +1은 시간 t를 위한 것
            nn.LeakyReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LeakyReLU(),
            nn.Linear(hidden_dim, input_dim)
        )

    def forward(self, x, t):
        # x: (batch, 784), t: (batch, 1)
        tx = torch.cat([x, t], dim=-1)
        return self.net(tx)
```

```python
def train_fm(model, dataloader, epochs=10):
    optimizer = optim.Adam(model.parameters(), lr=5e-4)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    for epoch in range(epochs):
        for x1, _ in dataloader:
            x1 = x1.view(-1, 784).to(device)
            x0 = torch.randn_like(x1) # 노이즈
            t = torch.rand(x1.shape[0], 1).to(device) # t ~ U(0, 1)

            # 직선 경로 생성
            xt = (1 - t) * x0 + t * x1
            target = x1 - x0 # 조건부 벡터장 u_t

            # Loss 계산 및 최적화
            v_theta = model(xt, t)
            loss = torch.mean((v_theta - target)**2)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# 데이터 로드 및 실행
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_loader = DataLoader(datasets.MNIST('.', train=True, download=True, transform=transform), batch_size=128, shuffle=True)
model = MLPVectorField()
train_fm(model, train_loader, epochs=100)
```

```python
@torch.no_grad()
def sample_fm(model, num_samples=10, steps=20):
    model.eval()
    device = next(model.parameters()).device
    x = torch.randn(num_samples, 784).to(device)
    dt = 1.0 / steps

    for i in range(steps):
        t = torch.ones(num_samples, 1).to(device) * (i / steps)
        v = model(x, t)
        x = x + v * dt # Euler step: x_{t+dt} = x_t + v*dt

    return x.view(-1, 28, 28).cpu()

samples = sample_fm(model)
plt.imshow(torch.cat([s for s in samples], dim=1), cmap='gray')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

# 1. 시간 임베딩 및 Conv 블록 정의
class TimeEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(1, dim),
            nn.SiLU(),
            nn.Linear(dim, dim)
        )
    def forward(self, t):
        return self.mlp(t)

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch, time_dim, apply_norm=True):
        super().__init__()
        self.conv = nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1)
        self.time_mlp = nn.Linear(time_dim, out_ch)
        self.apply_norm = apply_norm
        if apply_norm:
            # 채널이 8로 나누어 떨어지지 않으면 그룹을 1로 설정 (오류 방지)
            groups = 8 if out_ch % 8 == 0 else 1
            self.norm = nn.GroupNorm(groups, out_ch)
        self.act = nn.SiLU()

    def forward(self, x, t_emb):
        h = self.conv(x)
        # 시간 정보를 채널 차원에 더해줌 (B, C, 1, 1)
        h = h + self.time_mlp(t_emb)[:, :, None, None]
        if self.apply_norm:
            h = self.norm(h)
        return self.act(h)

# 2. CNN 기반 벡터장 신경망 (Vector Field Predictor)
class CNNVectorField(nn.Module):
    def __init__(self):
        super().__init__()
        self.t_embed = TimeEmbedding(64)
        self.down1 = ConvBlock(1, 64, 64)
        self.down2 = ConvBlock(64, 128, 64)
        self.mid = ConvBlock(128, 128, 64)
        self.up2 = ConvBlock(128, 64, 64)
        self.up1 = ConvBlock(64, 1, 64, apply_norm=False) # 마지막은 정규화 제외

    def forward(self, x, t):
        if t.shape[0] != x.shape[0]:
            t = t.expand(x.shape[0], 1)
        
        t_emb = self.t_embed(t)
        h1 = self.down1(x, t_emb)
        h2 = self.down2(h1, t_emb)
        h3 = self.mid(h2, t_emb)
        h4 = self.up2(h3, t_emb) + h1 # Skip connection
        out = self.up1(h4, t_emb)
        return out

# 3. 학습 함수
def train():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # 데이터 준비 (MNIST)
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,)) # -1 ~ 1 사이로 정규화
    ])
    dataset = datasets.MNIST('.', train=True, download=True, transform=transform)
    loader = DataLoader(dataset, batch_size=128, shuffle=True)

    model = CNNVectorField().to(device)
    optimizer = optim.Adam(model.parameters(), lr=5e-3)
    epochs = 100

    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for x1, _ in loader:
            x1 = x1.to(device)
            x0 = torch.randn_like(x1).to(device) # 노이즈
            t = torch.rand(x1.shape[0], 1).to(device) # t ~ U(0, 1)

            # Flow Matching 직선 경로: x_t = (1-t)x0 + t*x1
            # t의 차원을 x1에 맞게 확장 (B, 1, 1, 1)
            t_expand = t[:, :, None, None]
            xt = (1 - t_expand) * x0 + t_expand * x1
            
            # 타겟 벡터장: u_t = x1 - x0
            target = x1 - x0

            # 예측 및 손실 계산
            v_theta = model(xt, t)
            loss = torch.mean((v_theta - target)**2)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        
        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(loader):.4f}")
    
    return model

# 4. 샘플링 및 시각화 (Euler Method)
@torch.no_grad()
def sample(model, steps=50):
    model.eval()
    device = next(model.parameters()).device
    x = torch.randn(16, 1, 28, 28).to(device) # 16개 샘플 생성
    dt = 1.0 / steps

    for i in range(steps):
        t = torch.ones(16, 1).to(device) * (i / steps)
        v = model(x, t)
        x = x + v * dt # x_{t+dt} = x_t + v * dt

    # 결과 시각화
    x = x.cpu().numpy()
    plt.figure(figsize=(8, 8))
    for i in range(16):
        plt.subplot(4, 4, i+1)
        plt.imshow(x[i, 0], cmap='gray')
        plt.axis('off')
    plt.show()

# 실행
trained_model = train()
sample(trained_model)
```
